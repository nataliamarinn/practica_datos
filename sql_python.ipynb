# Lectura e informe de datos

```python
import sqlite3
import pandas as pd
```

```python
# Conexión a la base de datos
conn=sqlite3.connect("northwind.db")
```

```python
# Para ver las tablas vamos a usar una función de pandas que se llama pd.read_sql 
tablas= pd.read_sql("""SELECT name, type FROM sqlite_master WHERE type IN ('table','view') ORDER BY name""",conn)
print(tablas)
```

Table vs View

- Table: son los datos de las tablas originales
- View: es cuando se guarda una query especifica, tipo atajo

```python
# guardar una lista con los nombres de las tablas

lista_tablas= tablas[tablas["type"]=="table"]['name'].tolist()
#ordenar la lista, por defecto viene ordenada, pero si quisiera
lista_tablas.sort(reverse=False)
lista_tablas
```

Para ver las columnas de cada tabla, los tipos y los primary keys

PRAGMA table_info() -> devuelve una tabla con columnas que tienen nombres técnicos.
por defecto me devuelve la tabla con las columnas: cid, name, type, notnull, dflt_value,pk

```python
def ver_columnas(tabla):
    df = pd.read_sql(f'PRAGMA table_info("{tabla}")', conn)
    df.columns=['cid','nombre','tipo','not_null','default','es_pk'] #renombrando las columnas
    df['es_pk']=df['es_pk'].apply(lambda x: 'Si' if x==1 else 'No')
    df['not_null']= df['not_null'].apply(lambda x: 'NO Null' if x==1 else 'nullable')
    return df[['nombre','tipo','not_null','es_pk']]

for tabla in lista_tablas:
    print(f'\n {tabla.upper()}')
    print(ver_columnas(tabla).to_string(index=False))
```
